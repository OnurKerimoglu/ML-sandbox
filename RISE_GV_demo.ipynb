# Graphviz

```python
import graphviz
dot = graphviz.Digraph('2_phyto_1_resource')  
#nodes
dot.node('p1', r'P1')
dot.node('p2', r'P2')
dot.node('n', 'N')
#edges
dot.edge('p1', 'n', dir='back')
dot.edge('p2', 'n', dir='back')
dot
```
